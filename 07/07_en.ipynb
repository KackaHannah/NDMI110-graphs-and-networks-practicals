# The Giant Component

> **Theorem (Supercritical Erdős-Rényi graph)**
>
> Fix $c > 1$ and let $G_n \sim G(n, c/n)$. Let $C_1 = C_1(G_n)$ denote the largest connected component of $G_n$.
>
> Then there exists a unique number $\rho(c) \in (0, 1)$ satisfying
> $$
> \rho(c) = 1 - e^{-c\rho(c)},
> $$
> and
> $$
> \frac{|C_1|}{n} \to \rho(c)
> $$
> in probability as $n \to \infty$. Moreover, with high probability,
> $$
> |C_2| = o(n),
> $$
> where $C_2 = C_2(G_n)$ denotes the second largest connected component.

For $p = c/n$, the structure of the network depends on $c$.
According to **Theorem (Supercritical Erdős-Rényi graph)**, if $c > 1$, there exists a unique giant component $C_1$ containing a fraction of the nodes.
The asymptotic proportion of nodes in the giant component, $\rho(c) = |C_1|/n$, is given by the unique solution in $(0, 1)$ to the equation:
$$
\rho = 1 - e^{-c\rho}
$$

# Tree-like Local Exploration

In the lecture, it is noted that exploring a component using Breadth-First Search (BFS) in $G(n, c/n)$ behaves asymptotically like a **Galton-Watson branching process** with offspring distribution Poisson($c$).

The key insight is that when the explored set $S$ has size $s \ll n$, each newly discovered vertex has approximately $\mathrm{Bin}(n - s, c/n)$ new neighbors. Since $s/n \to 0$, this converges to $\mathrm{Poisson}(c)$. Crucially, this approximation is only valid while the explored set remains small. Specifically, the expected number of collisions (edges back into the already-visited set) when exploring up to $\log n$ nodes is:
$$
\mathcal{O}\!\left( \frac{(\log n)^2}{n} \right) = o(1)
$$
so the exploration is tree-like up to that scale, and the Poisson approximation holds.
